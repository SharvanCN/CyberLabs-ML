```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df_x = pd.read_csv('polynomial_train.csv')
```

```python
x_train = df_x.to_numpy()
y_train = x_train[:,4]
x_train = np.delete(x_train, [0,4] , axis = 1)
```

```python
x_train_mean = np.mean(x_train, axis = 0) 
x_train_std = np.std(x_train, axis = 0)
x_train = (x_train-x_train_mean)/x_train_std
n = 5
q = x_train[:, 0]
p = x_train[:, 1]
e = x_train[:, 2]
```

```python
model_polynomial = []

for i in range(n+1):
    for j in range(n+1):
        for k in range(n+1):
            if i+j+k <= n & i+j+k>0:
            
                term = (q**i) * (p**j) * (e**k)
                model_polynomial.append(term)

model_polynomial = np.array(model_polynomial).T
print(model_polynomial.shape)
```

```python
m,n = x_train.shape
w_initial = np.zeros(model_polynomial.shape[1])
b_initial = 0
y_hat = np.zeros(model_polynomial.shape[0])
dj_dw = np.zeros(model_polynomial.shape[1])
dj_db = 0
tmp_alpha = 0.1
iterations = 15000
lamda = 2
J_history = []
p_history = []
```

```python
x_traintrain = np.delete(model_polynomial, range(15000, 20000), axis  = 0)
nn = x_traintrain.shape[0]
x_traintest = model_polynomial[15000:20000 , :]
y_traintrain = np.delete(y_train, range(15000,20000), axis = 0)
y_traintest = y_train[15000:20000]
y_meanmean = np.sum(y_traintest)/100000
x_traintrain_mean = np.mean(x_traintrain, axis = 0)
x_traintrain_std = np.std(x_traintrain, axis = 0)
x_traintrain = (x_traintrain - x_traintrain_mean)/x_traintrain_std
x_traintest_mean = np.mean(x_traintest, axis = 0)
x_traintest_std = np.std(x_traintest, axis = 0)
x_traintest = (x_traintest - x_traintest_mean)/x_traintest_std
```

```python
def model(x, w, b):
    y_hat = np.dot(x, w) + b
    return y_hat
```

```python
def cost_func(y_hat, y, w, lamda):
    loss = np.square(y_hat - y)/2
    cost_1 = np.mean(loss)
    cost_2 = lamda*np.square(w)
    cost_3 = np.sum(cost_2/(2*model_polynomial.shape[1]))
    final_cost = cost_1 + cost_3
    return final_cost
```

```python
def grad(x, y, y_hat):
    dj_dw = (np.dot(x.T, (y_hat - y)))/m
    dj_db = np.sum(y_hat - y)/m
    return dj_dw, dj_db
```

```python
def gradient_descent(x, y, w, b,lamda, alpha, num_iters, cost_func, grad):
    w = np.zeros(model_polynomial.shape[1])
    for i in range(num_iters):
        y_hat = model(x, w, b)
        dj_dw, dj_db = grad(x, y, y_hat)     
        w -= alpha * dj_dw + (lamda*np.sum(w)/m)
        b -= alpha * dj_db 
        J_history.append(cost_func(y_hat, y, w, lamda))
        p_history.append([w,b])
    return w, b, J_history, p_history
```

```python
w_final, b_final, J_hist, p_hist = gradient_descent(x_traintrain, y_traintrain, w_initial, b_initial, lamda, tmp_alpha, 
                                                    iterations, cost_func, grad)
```

```python
%%time
print(J_hist, w_final, b_final)
```

```python
y_hat = model(x_traintrain, w_final, b_final)
rmse = np.sqrt(np.mean((y_traintrain - y_hat)**2))
print(f"RMSE: {rmse:.4f}")
```

```python
y_hat = model(x_traintest, w_final, b_final)
rmse = np.sqrt(np.mean((y_traintest - y_hat)**2))
print(f"RMSE: {rmse:.4f}")
```
